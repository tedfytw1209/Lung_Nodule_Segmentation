```python
# Directory containing DICOM files
dicom_dir = '/blue/bianjiang/imaging_study/IRB202201140_Lung_Cancer/img/arc001/IRB202201140_C_26146/SCANS/3/DICOM'

# Load DICOM images
dicom_files = [pydicom.dcmread(os.path.join(dicom_dir, f), force=True) for f in sorted(os.listdir(dicom_dir))]


# Assuming all DICOM files have the same dimensions
image_shape = list(dicom_files[0].pixel_array.shape)
image_shape.append(len(dicom_files))

# Create a 3D numpy array to hold the stacked images
image_volume = np.zeros(image_shape)

# Stack images
for i, dicom_file in enumerate(dicom_files):
    image_volume[:, :, i] = dicom_file.pixel_array

# Visualize a slice
plt.imshow(image_volume[:, :, image_volume.shape[2] // 2], cmap='gray')
plt.show()
```

```python
import os
import numpy as np
import pydicom
from pydicom.pixel_data_handlers.util import apply_modality_lut, apply_voi_lut
import matplotlib.pyplot as plt

def load_dicom(path):
    dicom = pydicom.dcmread(path, force=True)
    if dicom.pixel_array.ndim == 2:
        # Apply Modality LUT and VOI LUT for proper intensity transformations
        data = apply_modality_lut(dicom.pixel_array, dicom)
        data = apply_voi_lut(data, dicom)
        return data
    return None

dicom_dir = '/blue/bianjiang/imaging_study/IRB202201140_Lung_Cancer/img/arc001/IRB202201140_C_26146/SCANS/3/DICOM'
dicom_files = [f for f in sorted(os.listdir(dicom_dir)) if f.endswith('.dcm')]
image_volume = np.stack([load_dicom(os.path.join(dicom_dir, f)) for f in dicom_files])

# Visualize one slice
plt.imshow(image_volume[:, :, image_volume.shape[2] // 2], cmap='gray')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

def plot_slice(slice_index):
    plt.figure(figsize=(8, 8))
    plt.imshow(image_volume[:, :, slice_index], cmap='gray')
    plt.title(f'Slice {slice_index}')
    plt.axis('off')
    plt.show()

# Assuming 'image_volume' is your 3D numpy array
interact(plot_slice, slice_index=IntSlider(min=0, max=image_volume.shape[2] - 1, step=1, value=image_volume.shape[2] // 2));
```
